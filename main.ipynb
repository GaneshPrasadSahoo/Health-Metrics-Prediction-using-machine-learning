Encode The CategoricalData 

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Gender"]=le.fit_transform(df[["Gender"]])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["BMI Category"]=le.fit_transform(df[["BMI Category"]])
```

The Value Of Daily Step Is Very Large So We Do Scaling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df["Daily Steps"]=ss.fit_transform(df[["Daily Steps"]])
```

Separate Systolic and Diastolic Values

```python
# Assuming 'Blood Pressure' is a column in your DataFrame
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)

# Convert the new columns to numeric
df['Systolic'] = pd.to_numeric(df['Systolic'], errors='coerce')
df['Diastolic'] = pd.to_numeric(df['Diastolic'], errors='coerce')
```

```python
df
```

```python
x = df[['Gender', 'Age', 'Sleep Duration', 'Physical Activity Level', 'Daily Steps', 'BMI Category']]
y = df[['Systolic', 'Diastolic', 'Stress Level', 'Heart Rate']]
```

```python
print("Shape of X:", x.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

from sklearn.multioutput import MultiOutputRegressor
```

```python
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(model.score(X_train,y_train)*100)
print(model.score(X_test,y_test)*100)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
dt=DecisionTreeRegressor(max_depth=9)
dt.fit(X_train,y_train)
```

```python
y_predi=dt.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import mean_squared_error

mae = mean_absolute_error(y_test, y_predi)
r2 = r2_score(y_test, y_predi)
mse = mean_squared_error(y_test, y_predi)
print("Mean Squared Error:", mse)

print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
print(dt.score(X_train,y_train)*100)
print(dt.score(X_test,y_test)*100)
```

```python
for i in range (1,10):
  dt=DecisionTreeRegressor(max_depth=i)
  dt.fit(X_train,y_train)
  print(dt.score(X_train,y_train),dt.score(X_test,y_test),i)



```

```python
plt.figure(figsize=(30, 15))
plot_tree(dt, filled=True, feature_names=X_train.columns, fontsize=10)
plt.title('Decision Tree Visualization')
plt.show()
```

K-NN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
kn=KNeighborsRegressor(n_neighbors=1)
kn.fit(X_train,y_train)
```

```python
Y_pred=kn.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import mean_squared_error

mae = mean_absolute_error(y_test, Y_pred)
r2 = r2_score(y_test, Y_pred)
mse = mean_squared_error(y_test, Y_pred)
print("Mean Squared Error:", mse)

print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
print(kn.score(X_train,y_train)*100)
print(kn.score(X_test,y_test)*100)
```

```python
for i in range(1,30):
  kn=KNeighborsRegressor(n_neighbors=i)
  kn.fit(X_train,y_train)
  print(i,kn.score(X_train,y_train)*100,kn.score(X_test,y_test)*100)
```

```python
models = ['RandomForest', 'DecisionTree', 'KNN']

mse = [0.495194, 0.25, 5.41]

mae = [0.1768, 0.08444, 0.38333]

r_squared = [0.9835, 0.9857, 0.8143]

bar_width = 0.2

r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.figure(figsize=(10,6))

plt.bar(r1, mse, color='blue', width=bar_width, edgecolor='grey', label='Mean Squared Error')

plt.bar(r2, mae, color='green', width=bar_width, edgecolor='grey', label='Mean Absolute Error')

plt.bar(r3, r_squared, color='red', width=bar_width, edgecolor='grey', label='R-squared')

plt.xlabel('Models', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(models))], models)
plt.ylabel('Metric Values')
plt.title('Model Performance Comparison')

plt.legend()

plt.tight_layout()
plt.show()
```
